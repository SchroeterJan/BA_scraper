```python
sidejobs = pd.read_excel("files\\sidejobs.xlsx")
try:
    sidejobs.drop(columns=["Unnamed: 0"], axis=1, inplace=True)
except:
    pass
#dt.datetime.fromtimestamp(1495534675)
sidejobs["created_at"] = sidejobs["created"].apply(lambda x: dt.datetime.fromtimestamp(x))
#sidejobs["created"].astype(str).astype('datetime64[ns]')
```

```python
candidacies_mandates = pd.read_excel("files\\candidacies_mandates.xlsx")

#candidacies_mandates.drop(columns=["Unnamed: 0"], axis=1, inplace=True)
```

```python
cand_mandates_relevant_cols = ['label', 'fraction_membership1', 'politician', 'electoral_list', 'constituency', 'parliament_period']
```

```python
parliament_periods = pd.read_excel("files\\parliament_periods.xlsx", converters={'start_date_period':pd.to_datetime,'end_date_period':pd.to_datetime} )
parliament_periods.drop(columns=["Unnamed: 0"], axis=1, inplace=True)
parliament_periods.head()
```

```python
parl_period_relevant_cols = ["label", "parliament", "type", "start_date_period", "end_date_period"]
```

```python
mandates_parliaments_merge = candidacies_mandates[cand_mandates_relevant_cols].merge(parliament_periods[parl_period_relevant_cols], left_on="parliament_period", right_on="parliament", how="left")
mandates_parliaments_merge.head()
```

```python
mandates_parliaments = mandates_parliaments_merge[["label_x","fraction_membership1","politician", "constituency", "parliament_period", "type", "start_date_period", "end_date_period"]].copy()
```

```python
sidejobs_pp_1 = sidejobs.merge(mandates_parliaments[["label_x", "start_date_period", "end_date_period"]], left_on=["mandate_1"], right_on=["label_x"], how="left", suffixes=(None, '_1'))
sidejobs_pp_2 = sidejobs_pp_1.merge(mandates_parliaments[["label_x", "start_date_period", "end_date_period"]], left_on=["mandate_2"], right_on=["label_x"], how="left", suffixes=(None, '_2'))
sidejobs_pp_3 = sidejobs_pp_2.merge(mandates_parliaments[["label_x", "start_date_period", "end_date_period"]], left_on=["mandate_3"], right_on=["label_x"], how="left", suffixes=(None, '_3'))
```

```python
sidejobs_pp_3.columns
```

```python
sidejobs_merge = sidejobs_pp_3.loc[:,~sidejobs_pp_3.columns.str.startswith('label')].copy()
sidejobs_merge.head()
```

```python
def get_relevant_period(row):
    print(row["created_at"])
    print(row["start_date_period"])
    
    if row["created_at"] >= row["start_date_period"] and row["created_at"] <= row["end_date_period"]:
        
        return row["mandate_1"]
    elif row["created_at"] >= row["start_date_period_2"] and row["created_at"] <= row["end_date_period_2"]:
        return row["mandate_2"]
    elif row["created_at"] >= row["start_date_period_3"] and row["created_at"] <= row["end_date_period_3"]:
        return row["mandate_3"]
    else:
        return None
    
```

```python
sidejobs_merge['relevant_period'] = sidejobs_merge.apply(lambda x: get_relevant_period(x), axis=1)
```

```python
sidejobs_merge.merge(mandates_parliaments, left_on=["relevant_period"], right_on=["label_x"], how="left", suffixes=(None, '_rel'))
```
